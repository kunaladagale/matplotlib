```python
p = [26, 39, 43, 59, 44, 54, 45, 53, 35, 42, 38, 45, 16, 10, 41, 51, 40,
       55, 19, 39, 53, 26, 31, 23, 33, 49, 49, 33, 29, 11, 44, 10, 46, 51,
       13, 31, 50, 44, 12, 58, 34, 41, 11, 24, 18, 43, 14, 20, 54, 54]
plt.hist(p)
plt.show()
# basci histogram
```

```python
p = [26, 39, 43, 59, 44, 54, 45, 53, 35, 42, 38, 45, 16, 10, 41, 51, 40,
       55, 19, 39, 53, 26, 31, 23, 33, 49, 49, 33, 29, 11, 44, 10, 46, 51,
       13, 31, 50, 44, 12, 58, 34, 41, 11, 24, 18, 43, 14, 20, 54, 54]
plt.hist(p,color="g",edgecolor ="r")
plt.xlabel("language")
plt.ylabel("numbers")
plt.title("language and numbers popularity")
plt.show
# color parameter used to change the color of the histogram 
```

```python
l=[10,20,30,40,50,60]
plt.hist(p,color="g",bins = l,edgecolor ="r")
plt.xlabel("language")
plt.ylabel("numners")
plt.title("language and numbers popularity")
plt.show()
```

```python
#l=[10,20,30,40,50,60]
plt.hist(p,color="g",bins = 5,edgecolor ="r",histtype = "step")
plt.xlabel("language")
plt.ylabel("numners")
plt.title("language and numbers popularity")
plt.show()
# histtype parameter is used for changing the type of histogram 
# histtype{'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar'
```

```python
l=[10,20,30,40,50,60]
plt.hist(p,color="g",bins = l,edgecolor ="r",orientation = "horizontal")
plt.xlabel("language")
plt.ylabel("numners")
plt.title("language and numbers popularity")
plt.show()
# orientation{'vertical', 'horizontal'}, default: 'vertical'  
```

```python
l=[10,20,30,40,50,60]
plt.hist(p,color="g",bins = l,edgecolor ="r",label = "python")
plt.xlabel("language")
plt.ylabel("numners")
plt.title("language and numbers popularity")
plt.legend()
plt.grid()
plt.show()


# plt.grid = used for grid line 
```
