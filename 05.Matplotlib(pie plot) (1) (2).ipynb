# pie plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["c","c++","java","python"]
ex = [0.0,0.0,0.0,0.2]
c = ["b","g","r","y"]
plt.pie(x,labels= y,explode = ex,colors = c,autopct = "%0.0f%%",shadow = True)
plt.show()

# explode parameter is used for the taking this part little bit aside 
# autopct is used to display the percentage inside the pie chart
# shadow is used for the shadow of the pie chart
```

```python
x = [10,20,30,40]
y = ["c","c++","java","python"]
ex = [0.2,0.0,0.0,0.0]
c = ["b","g","r","y"]
plt.pie(x,labels= y,explode = ex,colors = c,autopct = "%0.1f%%",shadow = True,radius = 1,labeldistance = 1.2)
plt.title("pie chart")
plt.legend()
plt.show()

# radius is used to change the radius of the pie chart 
#labeldistance is used to change the distance of the label provided on pie chart 
```

# dot pie chart/plot

```python
plt.pie([1])
plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["c","c++","java","python"]



plt.pie(x,labels= y,radius = 1)
plt.pie(x1,radius = 0.75,colors = ["r","g","b","y"])

plt.title("pie chart")

plt.legend()
plt.show()
```

```python
x = [10,20,30,40]

y = ["c","c++","java","python"]



plt.pie(x,labels= y,radius = 1.3,labeldistance = 0.9)   # first pie chart
plt.pie([1],colors = "w")

plt.title("pie chart")

plt.legend()
plt.show()
```
