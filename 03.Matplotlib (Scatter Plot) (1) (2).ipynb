```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,2,6,3,8]

plt.scatter(x,y,color = "r")

plt.title("scatter plot",fontsize = 12)
plt.xlabel("day")
plt.ylabel("number")
plt.show()
# color used to change the color 
# fontsize used to change the font size
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,2,6,3,8]
color = [10,23,20,34,40,35,45]
plt.scatter(x,y,c = color,marker = "v",linewidth = 2,cmap = "rainbow",alpha = 0.9)

t = plt.colorbar()
t.set_label("Colorbar",fontsize = 20)
plt.title("scatter plot",fontsize = 20)
plt.xlabel("day")
plt.ylabel("number")
plt.show()
 
# marker parameter is used for change the shape of the points 
# linewidth = used for the changing the size of points 
# cmap = used for the to change the colour 
# colorbar is used to show the color bar of the cmap 
# alpha parameter is used for the lighting the colour 0-1
```
