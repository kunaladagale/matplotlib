# Stack Plot

```python
import matplotlib.pyplot as plt
```

stack plot is also called as area plot

```python
x = [1,2,3,4,5,6]
area = [2,2,3,4,5,-3]
area1=[4,6,7,8,-9,3]
area2=[7,8,9,2,1,5]
plt.stackplot(x,area,area1,area2,labels = ["area","area1","area2"],colors = ["r","g","b"],baseline = "zero")
plt.title("python")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()



# labels parameter used to give the labeling to the graph 
# colors parameter is used to change the colours 
# baseline parameter = "zero" == no change
#                       "sym"    == base line get hide 
#                    " wiggle"   == base line get hide 
```
