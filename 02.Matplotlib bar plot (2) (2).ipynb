# Bar Plot

```python
import matplotlib.pyplot as plt
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]

plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)

plt.bar(x,y,width = 0.8,color = "r")
plt.show()

# fontsize = parameter used to increase or decrease the size of xlabel,ylabel,title
# width = parameter is used to change the width of the bar  
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]

plt.title("language popularity",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.xlabel("language",fontsize = 10)


plt.bar(x,y,width = 0.8,color = "r")
plt.show()
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.bar(x,y,width = 0.4,color = ["r","b","m","g"])
plt.show()

# if we want different colours for the different bar graph then just pass the list of colours 
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.bar(x,y,width = 0.4,align = "center")
plt.show()
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.bar(x,y,width = 0.4,color = ["r","b","m","g"],align = "center")
plt.show()

#align parameter is used to the take the language word at the edge align contains two parameter "edge" and second one is "center "
```

b : blue.
g : green.
r : red.
c : cyan.
m : magenta.
y : yellow.
k : black.
w : white.

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.bar(x,y,width = 0.4,color = ["r","b","m","g"],align = "edge",edgecolor = "black",linewidth = 5,linestyle ="--",alpha=0.9)
plt.show()

# edgecolour parameter is used for taking the edges of the bar 
#linewidth is used for incresing the width of the edgecolor
#linestyle is used for changeing the styple of the line "-.","--","-"
# aplha parameter is used for lighting the colours of the bar range(0-1)
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.bar(x,y,width = 0.4,color = ["g","b","m","r"],align = "edge",label = ["python","c++","java","c"])
plt.legend()
plt.show()
# label is used for the give the lable inside the plot but for that we want to give the legend key 
```

```python
x =["python","c++","java","c"]
y =[85,70,80,71]
plt.xlabel("language",fontsize = 10)
plt.ylabel("popularity",fontsize = 10)
plt.title("language popularity",fontsize = 10)
plt.barh(x,y,color = ["r","b","m","g"])
plt.show()

# barh is used to make the bars in horizontal form 
```
