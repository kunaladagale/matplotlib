```python
 import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,100]
plt.boxplot(x,notch = False)
plt.show()
```

```python
x = [10,20,30,40,50,100]
plt.boxplot(x,notch = True)
plt.show()

# notch paramters used to change the shape of box 
```

```python
x = [10,20,30,40,50,100]
plt.boxplot(x,vert= False,widths = 0.2)
plt.show()

# vert parameter used to change vertical to horizontal 
# width parameters used to change the width of the box 
```

```python
x = [10,20,30,40,50,100]
plt.boxplot(x,labels = ["python"],patch_artist = True,showmeans=True,sym = "r*")
plt.show()

# leables parameter is used to give the name to the box plot it is passed always in a list formate
# patch_artist used to fill the box
# sym parameter used to change the shape and colour of the outlier
```

```python
x = [10,20,30,40,50]
plt.boxplot(x,labels = ["python"],showmeans=True)
plt.show()

# showmean used to show the mean 
```
